```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm

import os
import gc

```

```python
types = "clicksxcarts"

path_input_bpr = {}
path_output = {}
for i in ["train1", "valid"]:
    path_input_bpr[i] = {}
    for j in ["clicksxcarts", "cartsxorders", "ordersxclicks"]:
        path_input_bpr[i][j] = {}
        for k in ["session2idx", "user_items", "model", "list_aid"]:
            path_input_bpr[i][j][k] = f"/kaggle/input/otto-cf-bpr-{j}/{k}_{j}_{i}.pickle"
    path_output[i] = f"sim_items_bpr_{types}_{i}.pickle"
    

```

```python
# 特徴量を作成
for i in ["train1", "valid"]:  
    dict_recom_all = {}
    # BPR
    model = pd.read_pickle(path_input_bpr[i][types]["model"])
    session2idx = pd.read_pickle(path_input_bpr[i][types]["session2idx"])
    user_items = pd.read_pickle(path_input_bpr[i][types]["user_items"]).tocsr()
    list_aid = pd.read_pickle(path_input_bpr[i][types]["list_aid"])
    aid2idx = dict(zip(list_aid, range(len(list_aid))))
    
    dict_sim = {}
    for aid in tqdm(list_aid):
        sim = model.similar_items(aid2idx[aid], N = 5)
        dict_sim[aid] = [list_aid[pair[0]] for pair in sim[1:]]
    pd.to_pickle(dict_sim, path_output[i])
```
